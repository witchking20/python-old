```python
def page_down():
    global curr_height
    global prev_height
    
    prev_height = driver.execute_script('return document.body.scrollHeight')

    while True:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(3)

        curr_height = driver.execute_script('return document.body.scrollHeight')
        if curr_height == prev_height:
            break
        prev_height = curr_height
```

```python
def page_up():
    driver.execute_script("window.scrollTo(0, 0);")
```

```python
def page_down2():
    global page_loc
    global curr_height
    global done
    page_loc = page_loc + 200
    driver.execute_script(f'window.scrollTo(0, {page_loc});')
    time.sleep(3)
    prev_height = driver.execute_script('return document.body.scrollHeight')
    if curr_height == prev_height:
        print(f'{i-1}개의 데이터 수집완료')
        done = True
        return
    else:
        try:
            print(curr_height,'사번')
            driver.find_element(By.XPATH, value=f'//*[@id="__next"]/div[3]/div[2]/ul/li[{i}]/div/a/div[1]').click()
            
            driver.switch_to.window(driver.window_handles[0])
            craw()
            driver.back()
            driver.switch_to.window(driver.window_handles[0])
            return
        except:
            time.sleep(3)
            curr_height = driver.execute_script('return document.body.scrollHeight')
            print(curr_height,'오번')
            page_down2()
```

```python
def craw():
    for j in range(1,5):
        time.sleep(3)
        if j == 1:
            try:
                element = driver.find_element(By.CLASS_NAME, value = 'JobHeader_JobHeader__DBrFJ')
                name.append(element.text.split('\n')[0].split('∙')[0])
                loc.append(element.text.split('\n')[0].split('∙')[1])
                career.append(element.text.split('\n')[0].split('∙')[2])
                title.append(element.text.split('\n')[1])

                one.append(element.text.split('\n')[0].split('∙')[0])
                two.append(element.text.split('\n')[0].split('∙')[1])
                three.append(element.text.split('\n')[0].split('∙')[2])
                four.append(element.text.split('\n')[1])
            except:
                name.append('')
                loc.append('')
                career.append('')
                title.append('')
        elif j == 2:
            try:
                element = driver.find_element(By.CLASS_NAME, value = 'CompanyTags_CompanyTags__wpSCo')
                if len(element.text.split('\n')) >=1:
                    five.append(element.text.split('\n')[1:])
                    tag.append(element.text.split('\n')[1:])
                else:
                    five.append('')
                    tag.append('')    
            except:
                five.append('')
                tag.append('')
        elif j == 3:
            try:
                element = driver.find_element(By.CLASS_NAME, value = 'JobDueTime_JobDueTime__iKbEO')
                if len(element.text.split('\n')) ==2:
                    six.append(element.text.split('\n')[1])
                    timeout.append(element.text.split('\n')[1])
                else:
                    six.append('')
                    timeout.append('')
            except:
                six.append('')
                timeout.append('')
        else:
            try:
                element = driver.find_element(By.CLASS_NAME, value = 'JobWorkPlace_JobWorkPlace__5fs5x')
                if len(element.text.split('\n')) == 2:
                    seven.append(element.text.split('\n')[1])
                    loc2.append(element.text.split('\n')[1])
                else:
                    seven.append('')
                    loc2.append('')    
            except:
                seven.append('')
                loc2.append('')
```
